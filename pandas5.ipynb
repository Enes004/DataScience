```python
def salary_category(salary):
    if salary<50000:
        return "low"
    elif 50000<salary<80000:
        return "mid"
    else:
        return "high"
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
def per_score(score):
    if score>10:
        return 1
    else:
        return 0
```

```python
df["values"]= df["Experience"].apply(per_score)
```

```python
df["New_Score"] = df["values"]+df["Performance_Score"]
```
