```python
from abc import ABC,abstractmethod
```

```python
#Interface
class CanBeAttack(ABC):
    @abstractmethod
    def attack(self,target):
        pass
```

```python
#Abstract base Class
class Character(ABC):
    
    __slots__ = ["_name","_life","_level"] #memory optimization
    
    def __init__(self,name,life,level):
        self._name = name
        self._life = life
        self._level = 1

    
    
```
