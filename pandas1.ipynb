```python
import numpy as np
import pandas as pd
```

```python
#Series
#Series is a one-dimensional labeled data structure that can hold any data type (integers, strings, floats, etc.).
```

```python
grades = {"enes":1 , "ali":2 , "said":3}
pd.Series(grades)
```

```python
grades = [5,10,15,20]
students = ["enes","ali","said","veli"]
pd.Series(data=grades , index = students)
```

```python
#Series creation 
contest_result1 = pd.Series(data=[10,20,30],index=["enes","ali","said"])
contest_result2 = pd.Series(data=[85,45,37],index=["enes","ali","said"])
```

```python
contest_result1["enes"]
```

```python
contest_result2["ali"]
```

```python
whoiswinner = contest_result1 + contest_result2
```

```python
whoiswinner.idxmax()
```

```python
whoiswinner
```

```python
whoiswinner = contest_result1 * contest_result2
whoiswinner
```

```python
whoiswinner = contest_result1 / contest_result2
whoiswinner
```
