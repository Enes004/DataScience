```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,12,18,21,28]
weight_list = [60,70,76,87,82]
```

```python
plt.plot(age_list,weight_list,"y")
plt.xlabel("age")
plt.ylabel("weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"g*-")
plt.xlabel("age")
plt.ylabel("weight")
plt.show()
```

```python
np1 = np.linspace(0,10,20)
np1
```

```python
np2 = np1**3
```

```python
plt.plot(np1,np2,"r*-")
plt.show()
```

```python
plt.plot(np2,np1,"b--")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(np2,np1,"b--")
plt.subplot(1,2,2)
plt.plot(np1,np2,"r*-")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(np1,np2,"r--")
plt.subplot(2,1,2)
plt.plot(np2,np1,"g-*")
plt.show()
```

```python
my_figure = plt.figure(dpi=100)
figure_axes = my_figure.add_axes([0.1,0.1,0.6,0.6]) #size
figure_axes.plot(np1,np2,"g*-")
figure_axes.set_xlabel("x ekseni")
figure_axes.set_ylabel("y ekseni")
figure_axes.set_title("Baslik")
plt.show()
```

```python
figure = plt.figure(dpi=100)
axes = figure.add_axes([0.1,0.1,0.9,0.9]) #size
axes.plot(np1,np2**4,"g*-" ,label="1")
axes.plot(np1,np2**3,"r*-", label="2")
axes.legend(loc=2)
plt.show()
```
