```python
class BattleBot: #my class structure
    def __init__(self,name,color,power,hp):  #constructor method
        self.name = name
        self.color = color
        self.power = power
        self.__battery = 100  #private
        self.__hp = hp  #Private

    def attack(self):
        if self.__battery > 0:
            print(f"{self.name} is attacking. Power is {self.power}")
            self.__battery -= 10
        else:
            print(" {self.name} 's battery is low , can't attack")

    def be_charged(self):
        print(f"{self.name} is charging")
        self.__battery = 100

    def take_damage(self,damage):
        if self.__hp-damage > 0:
            self.hp -= damage
            print(f"Robot got a damage , hp is {self.__hp}")
        else:
            self.__hp = 0
            print(f"robot has died")

    @property          #the pythonic shape of getter/setter. We can access and control hp without calling a function. 
    def hp(self):
        return self.__hp 

    @hp.setter
    def hp(self,new_value):
        if new_value < 0:
            self.__hp = 0
        elif new_value > 100:
            self.__hp = 100
        else:
            self.__hp = new_value
```
