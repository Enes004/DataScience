```python
#merge
```

```python
merge_df1 = pd.read_csv("7-merge_data1.csv")
```

```python
merge_df2 = pd.read_csv("7-merge_data2.csv")
```

```python
merge_df1
```

```python
merge_df2
```

```python
# merge outer-join -> bütün dataları ele alır , boş değerlere nan atar
```

```python
df_merge_outer = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="outer")
```

```python
df_merge_outer
```

```python
# merge left-join -> soldaki df ye göre elemanları seçer ve birleştirir
df_merge_left = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="left")
df_merge_left
```

```python
# merge right-join -> sağdaki df ye göre elemanları seçer ve birleştirir
df_merge_right = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="right")
df_merge_right
```

```python
# merge inner-join -> sadece ortak olan elemanları alıp birleştirir
df_merge_inner = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="inner")
df_merge_inner
```
