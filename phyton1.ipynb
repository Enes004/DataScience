```python
def sesli_bul():
    sayac = 0
    sesliler = "aeiuüoöı"
    metin = input("metin giriniz")
    for harf in metin:
        for sesli in sesliler:
            if harf == sesli:
                sayac+=1
    return sayac

sesli_bul()
        
```
